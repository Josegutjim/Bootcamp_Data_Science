```python
df_limpio=df.copy()
```

```python
df_limpio['director'].fillna('Desconocida', inplace=True)
df_limpio['cast'].fillna('Desconocida', inplace=True)
df_limpio['country'].fillna('Desconocida', inplace=True)
```

```python
print(df_limpio.isnull().sum())
```

```python
df_limpio['date_added'] = pd.to_datetime(df_limpio['date_added'], errors='coerce')
df_limpio['date_added'] = df_limpio['date_added'].dt.strftime('%d/%m/%Y')
```

```python
df_limpio.head()
```

```python
#Películas vs Series
print(df_limpio['type'].value_counts())
```

```python
#Países con más contenido
from collections import Counter

all_countries = []
for countries in df_limpio['country']:
    if pd.notna(countries) and countries != 'Desconocida':
        all_countries.extend(countries.split(', '))

print(Counter(all_countries).most_common(10))
```

```python
#Años con más estrenos
df_years = df_limpio['release_year'].value_counts().reset_index()
df_years.columns = ['Año', 'Cantidad']
df_years = df_years.sort_values('Cantidad', ascending=False).head(10) 
print(df_years)
```

```python
#raiting de las peliculas
df_rating = df_limpio['rating'].value_counts().reset_index()
df_rating.columns = ['rating', 'Cantidad']
df_rating = df_rating.sort_values('Cantidad', ascending=False).head(10) 
print(df_rating)
```

```python
#Director con mas estrenos
df_director = df_limpio['director'].value_counts().reset_index()
df_director.columns = ['Director', 'Cantidad']
df_director=df_director[df_director['Director'] != 'Desconocida']
df_director = df_director.sort_values('Cantidad', ascending=False).head(10) 
print(df_director)


```

```python
df_limpio['type'].value_counts().plot(kind='bar')
plt.title('Películas vs Series en Netflix')
plt.ylabel('Cantidad')
plt.show()
```

```python
top_10 = Counter(all_countries).most_common(10)

df_top_countries = pd.DataFrame(top_10, columns=['País', 'Cantidad'])

plt.figure(figsize=(10,6))
plt.bar(df_top_countries["País"], df_top_countries["Cantidad"],color='g')
plt.title('Top 10 Países con más Contenido en Netflix')
plt.xticks(rotation=45)
plt.show()


```

```python
plt.figure(figsize=(14, 6))
plt.bar(df_years['Año'], df_years['Cantidad'], color='b')
plt.title('Contenido de Netflix por Año de Lanzamiento')
plt.xlabel('Año')
plt.ylabel('Cantidad')

plt.show()
```

```python
sns.barplot(x="rating", y="Cantidad", data=df_rating)
plt.title("Clasificación de Contenido en Netflix")
```

```python
sns.barplot(data=df_director, x='Cantidad', y='Director', color='purple')
plt.title('Top 10 Directores con más Contenido en Netflix')
plt.show()
```
