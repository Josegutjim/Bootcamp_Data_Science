Parte 2 — Limpieza de datos (Pandas)

```python
df1 = df.copy()
```

```python

# Corregir tipos de datos incorrectos
df1 ["cantidad"] = df1["cantidad"].replace("three",3)

cols = ["precio", "cantidad"]

# 1. Reemplazamos espacios en blanco por NaN de numpy
df1[cols] = df1[cols].replace(r'^\s*$', np.nan, regex=True)

# 2. Convertimos los NaN en 0
df1[cols] = df1[cols].fillna(0)

# 3. (Opcional) Asegurar que sean números y no texto
df1[cols] = df1[cols].apply(pd.to_numeric)

print("Nulos después de la limpieza:\n", df1.isnull().sum())
```

```python
df1['pais'] = df1['pais'].replace({
    'peru':'Perú',
    'COL':'Colombia',
    'col':'Colombia',
    'chile':'Chile'
})
print("Valores únicos detectados en la columna 'pais':", df1['pais'].unique())

```

```python
df1['metodo_pago'] = df1['metodo_pago'].replace({
    'TRANSFERENCIA':'Transferencia',
    'transferencia':'Transferencia'
})
print("Valores únicos detectados en la columna 'metodo_pago':", df1['metodo_pago'].unique())
```

```python
#corregir el formato de fecha
df1['fecha'] = pd.to_datetime(df1['fecha']).dt.strftime('%d/%m/%Y')
```

Parte 3 — Análisis con Pandas

```python

df1['precio'] = pd.to_numeric(df1['precio'], errors='coerce')
df1['cantidad'] = pd.to_numeric(df1['cantidad'], errors='coerce')

df1['total'] = df1['precio'] * df1['cantidad']
```

```python
#Total de las ventas
Total_ventas = df1['total'].sum()
print("Total de las ventas:", Total_ventas)
```

```python
#Promedio de ventas
Promedio_ventas = df1['total'].mean()
print("Promedio de ventas:", Promedio_ventas)
```

```python
#Venta Maxima
Venta_maxima = df1['total'].max()
print("Venta maxima:", Venta_maxima)
```

```python
#Venta Minima
Venta_minima = df1['total'].min()
print("Venta minima:", Venta_minima)
```

Parte 4 — NumPy

```python
total_ventas=df1[['precio','cantidad']].to_numpy()

precios=total_ventas[:,0]
cantidad=total_ventas[:,1]
total=precios*cantidad
print(total_ventas)
print("Total de ventas", np.nansum(total))
```

Parte 5 — Análisis con NumPy

```python
print("Suma total de ventas:", np.nansum(total))
print("Promedio de ventas:", np.nanmean(total))
print("Venta maxima:", np.nanmax(total))
print("Venta minima:", np.nanmin(total))
print("Cantidad de ventas superirores a 1000:",len(total[total>10000]))
```

Parte 6 — Interpretación de resultados

1. ¿Los resultados obtenidos tienen sentido?
En general, sí tienen sentido.
Se ve una lógica clara en los datos, por ejemplo el cálculo de ventas (precio * cantidad) funciona correctamente, y también hay variedad de productos, clientes y países, lo cual es normal en un dataset de ventas.
Además, hay ventas tanto altas como bajas, lo que también es esperable.
Sin embargo, no se pueden considerar completamente confiables, porque hay varios problemas en la calidad de los datos que afectan los resultados.

2. ¿Detectaste valores sospechosos?
Sí, hay varios casos que llaman la atención.
Por ejemplo, aparecen precios extremadamente altos como 999999, que claramente no son reales. Este tipo de valores distorsiona completamente los resultados, especialmente el total vendido y el promedio.
También hay problemas en la cantidad, donde algunos registros tienen texto como “three” en lugar de números, o simplemente están vacíos. Esto puede generar errores o cálculos incorrectos.
Además, se encontraron valores nulos en columnas importantes como precio y cantidad, lo que afecta directamente los análisis.
Otro problema es la falta de estandarización en los textos. Por ejemplo, el país aparece de diferentes formas como “col”, “COL” o “Colombia”, y lo mismo pasa con el método de pago. Esto hace que al agrupar los datos se puedan duplicar categorías sin darse cuenta.
Finalmente, algunas fechas también se ven inconsistentes o repetidas, lo que puede afectar cualquier análisis temporal.
En resumen, el dataset tiene bastante “ruido”, algo muy común cuando los datos no han sido limpiados previamente.

3. ¿El promedio representa correctamente los datos?
No realmente.
El promedio se ve afectado por los valores extremos, como los precios exageradamente altos. Con solo uno de estos valores, el promedio puede subir mucho y dejar de representar la realidad.
También influye el hecho de que hay datos faltantes o incorrectos.
Por eso, sería más recomendable usar la mediana o calcular el promedio después de limpiar los datos y eliminar los outliers.

4. ¿Qué decisiones tomarías si esto fuera información real de negocio?
Lo primero que haría sería no tomar decisiones con los datos así como están.
Antes que nada, haría una limpieza completa:

corregir tipos de datos
manejar valores nulos
eliminar o ajustar errores
estandarizar columnas de texto

Luego revisaría los valores atípicos, especialmente los precios exagerados, para eliminarlos o validarlos con el área de negocio.
También implementaría reglas de validación, como asegurarse de que el precio sea mayor a cero, que la cantidad sea numérica y que las categorías como país estén controladas.
Después de esto, recalcularía todos los indicadores para tener resultados confiables.
Con datos limpios, ya sí se podrían sacar conclusiones reales, como identificar los productos más vendidos, analizar el comportamiento por país o evaluar los métodos de pago.
